## Comprensi√≥n de los Datos
Se realiza el analisis del dato y el contenido que tiene cada dataset de IEES

```python
# Instalar las librerias necesarias para el trabajo
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Encontrar todos los archivos que siguen el patr√≥n
file_pattern = 'data*.csv'
files = glob.glob(file_pattern)
files.sort()  # Ordenar los archivos si es necesario
```

```python
# Crear una lista para almacenar los dataframes
dataframes = []
```

```python
# Leer y mostrar cada archivo
for file in files:
    df = pd.read_csv(file)
    dataframes.append(df)
    print(f"Contenido de {file}:")
    print(df.head())  # Mostrar las primeras filas del dataframe
    print("\n")  # Espacio entre impresiones de archivos
```

```python
# Unir todos los DataFrames en uno solo
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
# Mostrar las primeras filas del DataFrame combinado
print("DataFrame combinado:")
print(combined_df.head())
```

```python
# Guardar el DataFrame combinado en un archivo CSV
combined_df.to_csv('dataset_unidos.csv', index=False)

print("El DataFrame combinado se ha guardado en 'dataset_unidos.csv'.")
```
